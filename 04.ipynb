```python
def cifrar(texto, n):
    resultado = []
    
    for caracter in texto:
        if 'a' <= caracter <= 'z':  # Letras minúsculas
            nuevo_caracter = chr((ord(caracter) - ord('a') + n) % 26 + ord('a'))
        elif 'A' <= caracter <= 'Z':  # Letras mayúsculas
            nuevo_caracter = chr((ord(caracter) - ord('A') + n) % 26 + ord('A'))
        elif '0' <= caracter <= '9':  # Dígitos
            nuevo_caracter = chr((ord(caracter) - ord('0') + n) % 10 + ord('0'))
        else:
            nuevo_caracter = caracter  # Otros caracteres no se modifican
        
        resultado.append(nuevo_caracter)
    
    return ''.join(resultado)

# Solicitar al usuario que ingrese un texto y el valor de n
texto = input("Ingresa el texto a cifrar: ")
n = int(input("Ingresa el valor de n: "))

# Cifrar el texto
texto_cifrado = cifrar(texto, n)

# Mostrar el resultado
print(f"\nTexto cifrado: {texto_cifrado}")
```

Pseudocodigo:
Función cifrar(texto, n):
    Crear una lista vacía llamada resultado
    
    Para cada caracter en texto:
        Si caracter es una letra minúscula ('a' <= caracter <= 'z'):
            Calcular nuevo_caracter como el desplazamiento del caracter dentro del rango de las letras minúsculas utilizando n
        Sino si caracter es una letra mayúscula ('A' <= caracter <= 'Z'):
            Calcular nuevo_caracter como el desplazamiento del caracter dentro del rango de las letras mayúsculas utilizando n
        Sino si caracter es un dígito ('0' <= caracter <= '9'):
            Calcular nuevo_caracter como el desplazamiento del caracter dentro del rango de los dígitos utilizando n
        Sino:
            Asignar nuevo_caracter como el caracter original (otros caracteres no se modifican)
        
        Añadir nuevo_caracter a la lista resultado
    
    Convertir la lista resultado en una cadena de texto y retornar el resultado
