```python
def es_palindromo(cadena):
    # Convertimos la cadena a minúsculas y eliminamos los espacios para asegurar una comparación precisa
    cadena = cadena.lower().replace(" ", "")
    
    # Verificamos si la cadena es igual a su reverso
    return cadena == cadena[::-1]

# Solicitamos al usuario que ingrese una cadena
cadena = input("Introduce una cadena: ")

# Comprobamos si la cadena es un palíndromo
if es_palindromo(cadena):
    print(f"La cadena '{cadena}' es un palíndromo.")
else:
    print(f"La cadena '{cadena}' no es un palíndromo.")
```

Función es_palindromo(cadena):
    Convertir la cadena a minúsculas y eliminar los espacios

    Si la cadena es igual a su reverso:
        Retornar Verdadero
    De lo contrario:
        Retornar Falso
Fin de la función

Solicitar al usuario que ingrese una cadena y almacenar en variable cadena

Si la función es_palindromo con la cadena dada retorna Verdadero:
    Imprimir que la cadena es un palíndromo
De lo contrario:
    Imprimir que la cadena no es un palíndromo
