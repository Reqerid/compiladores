```python
def es_alfabetica(palabra):
    # Convertimos la palabra a minúsculas para asegurarnos de que la comparación no sea sensible a mayúsculas
    palabra = palabra.lower()
    
    # Verificamos si la palabra está ordenada alfabéticamente
    if palabra == ''.join(sorted(palabra)):
        return True
    else:
        return False

# Solicitamos al usuario que ingrese una palabra
palabra = input("Introduce una palabra: ")

# Comprobamos si la palabra es alfabética
if es_alfabetica(palabra):
    print(f"La palabra '{palabra}' es alfabética.")
else:
    print(f"La palabra '{palabra}' no es alfabética.")
```

Función es_alfabetica(palabra):
    Convertir la palabra a minúsculas para evitar diferencias entre mayúsculas y minúsculas

    Si la palabra es igual a la palabra ordenada alfabéticamente:
        Retornar Verdadero
    De lo contrario:
        Retornar Falso
Fin de la función

Solicitar al usuario que ingrese una palabra y almacenar en variable palabra

Si la función es_alfabetica con la palabra dada retorna Verdadero:
    Imprimir que la palabra es alfabética

    Imprimir que la palabra no es alfabética
