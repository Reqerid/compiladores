```python
import math

def calcular_distancia(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def encontrar_punto_mas_cercano(p1, puntos):
    distancia_minima = float('inf')
    punto_mas_cercano = None

    for i, punto in enumerate(puntos):
        distancia = calcular_distancia(p1[0], p1[1], punto[0], punto[1])
        if distancia < distancia_minima:
            distancia_minima = distancia
            punto_mas_cercano = (i + 1, punto)  # Guardamos el ndice y el punto

    return punto_mas_cercano, distancia_minima

# Solicitar las coordenadas de los cinco puntos
punto1 = tuple(map(float, input("Ingresa las coordenadas del primer punto (x1, y1): ").split()))
punto2 = tuple(map(float, input("Ingresa las coordenadas del segundo punto (x2, y2): ").split()))
punto3 = tuple(map(float, input("Ingresa las coordenadas del tercer punto (x3, y3): ").split()))
punto4 = tuple(map(float, input("Ingresa las coordenadas del cuarto punto (x4, y4): ").split()))
punto5 = tuple(map(float, input("Ingresa las coordenadas del quinto punto (x5, y5): ").split()))

# Comparar los cuatro ltimos puntos con el primero
puntos = [punto2, punto3, punto4, punto5]
punto_mas_cercano, distancia_minima = encontrar_punto_mas_cercano(punto1, puntos)

# Mostrar el resultado
print(f"\nEl punto ms cercano al primero ({punto1}) es el punto {punto_mas_cercano[0]} {punto_mas_cercano[1]} con una distancia de {distancia_minima:.2f}")
```
