```python
def desglosar_cantidad(cantidad):
    # Definimos los valores de los billetes y monedas
    billetes = [500, 200, 100, 50, 20]
    monedas = [10, 5, 2, 1]

    print(f"Desglose de ${cantidad:.2f}:")

    # Desglosar en billetes
    for billete in billetes:
        if cantidad >= billete:
            num_billetes = cantidad // billete
            cantidad -= num_billetes * billete
            print(f"{int(num_billetes)} billete(s) de {billete} pesos")

    # Desglosar en monedas
    for moneda in monedas:
        if cantidad >= moneda:
            num_monedas = cantidad // moneda
            cantidad -= num_monedas * moneda
            print(f"{int(num_monedas)} moneda(s) de {moneda} peso(s)")

# Ejemplo de uso
desglosar_cantidad(434)
```
